```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

file = "s3://otus-fraud-suv/raw/2019-08-22.txt"

storage_options = {
    "client_kwargs": {
        "endpoint_url": "https://storage.yandexcloud.net"
    }
}

columns = [
    "transaction_id",
    "tx_datetime",
    "customer_id",
    "terminal_id",
    "tx_amount",
    "tx_time_seconds",
    "tx_time_days",
    "tx_fraud",
    "tx_fraud_scenario"
]

chunksize = 500_000

writer = None

for chunk in pd.read_csv(
    file,
    sep=",",
    names=columns,
    skiprows=1,
    dtype=str,
    storage_options=storage_options,
    chunksize=chunksize
):


    chunk["tx_datetime"] = pd.to_datetime(chunk["tx_datetime"], errors="coerce")

    num_cols = [
        "transaction_id",
        "customer_id",
        "terminal_id",
        "tx_amount",
        "tx_time_seconds",
        "tx_time_days",
        "tx_fraud",
        "tx_fraud_scenario"
    ]

    for c in num_cols:
        chunk[c] = pd.to_numeric(chunk[c], errors="coerce")

    chunk = chunk.drop_duplicates()

    chunk = chunk.dropna(subset=[
        "transaction_id",
        "tx_datetime",
        "customer_id",
        "terminal_id",
        "tx_amount"
    ])

    chunk = chunk[
        (chunk["tx_amount"] >= 0) &
        (chunk["tx_fraud"].isin([0, 1])) &
        (chunk["tx_fraud_scenario"] >= 0)
    ]
    chunk = chunk.reset_index(drop=True)

    table = pa.Table.from_pandas(chunk, preserve_index=False)


    if writer is None:
        writer = pq.ParquetWriter(
            "clean_transactions.parquet",
            table.schema
        )

    writer.write_table(table)

if writer:
    writer.close()

print("DONE: clean_transactions.parquet")
```

```python
storage_options = {
    "client_kwargs": {
        "endpoint_url": "https://storage.yandexcloud.net"
    }
}

df = pd.read_parquet("clean_transactions.parquet")

df.to_parquet(
    "s3://otus-fraud-suv/clean/clean_transactions.parquet",
    index=False,
    storage_options=storage_options
)

print("Uploaded to S3")
```
